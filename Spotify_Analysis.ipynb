```python
## importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Creation of dataframes by reading the csv file using read_csv function
df_tracks=pd.read_csv('E:/Spotify datasets/tracks.csv')
## head function shows first five rows if no parameter is passed
df_tracks.head(3)
```

```python
## Checking for null values in dataset
df_tracks.isnull().sum()
```

```python
## To get total no of rows and columns in dataset
df_tracks.info()
```

```python
## 10 least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending= True).head(10)
sorted_df
```

```python
## Descriptive Statistics on dataset
df_tracks.describe().transpose()
```

```python
## top 10 most popular songs with popularity greater than 90
## inplace is made as false so original data frame isn't changed
most_popular= df_tracks.query('popularity>90', inplace =False).sort_values('popularity', ascending= False)
most_popular[:10]
```

```python
## Setting Index on column release date, inplace=true means needs to change original dataframe

df_tracks.set_index('release_date',inplace=True)
## changing the value of indexed column to date time format
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
## Check Artist and name of song at 18 row in our dataset
df_tracks[['artists','name']].iloc[18]
```

```python
## Convert duration in milisecs to seconds, axis 0 implies rows and 1 means columns
df_tracks["duration"]=df_tracks['duration_ms'].apply(lambda x : round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
df_tracks.head()
```

```python
## Visualizations Corelation Map, annot true writes data value in each cell
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
heatmap.set_title("Correlation Heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
##Creating sample Data Frame and taking only .4% data
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
                          
```

```python
## Regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y ='loudness' ,x='energy', color='b').set(title="Loudnesss VS Energy Correlation")
```

```python
## Regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y ='popularity' ,x='acousticness', color='c').set(title="Popularity VS Acousticness Correlation")
```

```python
## creating a new column year
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
#
```

```python
total_dr=df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims =(10,5)
fig, ax =plt.subplots(figsize= fig_dims)
fig= sns.lineplot(x= years, y =total_dr, ax =ax).set(title='Year vs Duration')
plt.xticks(rotation=60)
```

```python
df_genre=pd.read_csv('E:/Spotify datasets/SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title("Duration of songs in different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in MilliSecs")
plt.ylabel("Genres")
plt.xticks(rotation=90)
```

```python
## Top 5 Genres by popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous= df_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title='Top 5 Genres by Popularity')
```
